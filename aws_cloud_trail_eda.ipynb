Triming AWS Cloud Trail Dataset

```python
import pandas as pd

FILE_PATH = "dec12_18features.csv"

chunk_size = 100000   # 100k rows at a time
chunks = []

selected_columns = [
    "eventID", "eventTime", "sourceIPAddress", "userAgent",
    "eventName", "eventSource", "awsRegion", "eventVersion",
    "userIdentitytype", "eventType", "userIdentityaccountId",
    "userIdentityprincipalId", "userIdentityarn",
    "userIdentityaccessKeyId", "userIdentityuserName",
    "errorCode", "errorMessage", "requestParametersinstanceType"
]

for chunk in pd.read_csv(FILE_PATH, chunksize=chunk_size, usecols=selected_columns):
    # Convert timestamps
    chunk["eventTime"] = pd.to_datetime(chunk["eventTime"], errors="coerce")
    
    # Append processed chunk
    chunks.append(chunk)

aws_df = pd.concat(chunks, ignore_index=True)
```

```python
aws_df.info()
print("First Few Entries\n")
aws_df.head(15)
print("Last Few Entries\n")
aws_df.tail(15)
```

Normalizing CloudTrail Dataset

```python
def normalize_aws_big(df):
    out = pd.DataFrame()

    out["timestamp"] = df["eventTime"]

    out["actor"] = df["userIdentityuserName"].fillna(
        df["userIdentityarn"]
    )

    out["source_ip"] = df["sourceIPAddress"]
    out["region"] = df["awsRegion"]
    out["user_identity"]=df["userIdentitytype"]
    out["event_name"] = df["eventName"]

    # status detection
    out["status"] = out["event_name"].apply(
        lambda x: "FAILED" if "Fail" in str(x) or "Denied" in str(x) else "SUCCESS"
    )
    # explicit errorCode overrides
    out["status"] = df["errorCode"].apply(
        lambda x: "FAILED" if pd.notna(x) else "SUCCESS"
    )

    # Include eventID for reference/debugging
    out["eventID"] = df["eventID"]

    return out


aws_norm = normalize_aws_big(aws_df)
aws_norm.head()
```

Threat scoring mechanism

```python
HIGH_RISK_EVENTS = [
    "CreateUser", "CreateAccessKey", "AttachRolePolicy",
    "PutUserPolicy", "AddUserToGroup", "UpdateAssumeRolePolicy",
]

def threat_score(row):
    score = 0

    # Failed actions
    if row["status"] == "FAILED":
        score += 5

    # High-risk events
    if any(evt in str(row["event_name"]) for evt in HIGH_RISK_EVENTS):
        score += 10

    # Suspicious region
    if row["region"] not in ["ap-south-1", "us-east-1"]:
        score += 3

    # Actor unknown
    if row["actor"] is None or row["actor"] == "":
        score += 2

    return score


aws_norm["threat_score"] = aws_norm.apply(threat_score, axis=1)
aws_norm.tail()
```

Success vs failed

```python
import matplotlib.pyplot as plt
import seaborn as sns

aws_norm["status"].value_counts()
plt.figure(figsize=(6,4))
aws_norm["status"].value_counts().plot(
    kind="bar",
    title="Success vs Failed Events in AWS CloudTrail",
)
plt.xlabel("Status")
plt.ylabel("Number of Events")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

Timeline of Events (Daily Trend)

```python
daily_counts = (
    aws_norm.set_index("timestamp")["eventID"]
            .resample("D")
            .count()
)

plt.figure(figsize=(14,4))
daily_counts.plot()
plt.title("Daily Event Volume in AWS CloudTrail")
plt.xlabel("Date")
plt.ylabel("Number of Events")
plt.grid(True)
plt.tight_layout()
plt.show()
```

FAILED Events Over Time

```python
failed_daily = (
    aws_norm[aws_norm["status"]=="FAILED"]
    .set_index("timestamp")["eventID"]
    .resample("D")
    .count()
)

plt.figure(figsize=(14,4))
failed_daily.plot(color='red')
plt.title("Daily Failed Events (Security-Relevant)")
plt.xlabel("Date")
plt.ylabel("Failed Events")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
aws_norm.groupby("actor")["threat_score"].sum().sort_values(ascending=False).head(10)
```

High-Risk Event Types (IAM Related)

```python
high_risk_mask = aws_norm["event_name"].isin(HIGH_RISK_EVENTS)
dangerous_events = aws_norm[high_risk_mask]["event_name"].value_counts()

plt.figure(figsize=(10,4))
dangerous_events.plot(kind="bar", color="orange")
plt.title("High-Risk IAM Events (Potential Attack Indicators)")
plt.xlabel("Event Name")
plt.ylabel("Count")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

Top Suspicious Actors (by Threat Score)

```python
top_actors = (
    aws_norm.groupby("actor")["threat_score"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure(figsize=(10,4))
top_actors.plot(kind="bar", color="purple")
plt.title("Top 10 Actors by Threat Score")
plt.xlabel("Actor")
plt.ylabel("Threat Score")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

Top Suspicious IPs

```python
top_ips = (
    aws_norm.groupby("source_ip")["threat_score"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure(figsize=(10,4))
top_ips.plot(kind="bar", color="brown")
plt.title("Top 10 Suspicious Source IPs")
plt.xlabel("IP Address")
plt.ylabel("Threat Score")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

Heatmap  Region vs Status

```python
import seaborn as sns

heatmap_df = pd.crosstab(
    aws_norm["region"],
    aws_norm["status"]
)

plt.figure(figsize=(10,6))
sns.heatmap(heatmap_df, cmap="Reds", linewidths=0.5)
plt.title("Region vs Status Heatmap")
plt.tight_layout()
plt.show()
```
